```python
# ============================================
# PROGRAMA IA BASICO SOBRE IBAN
# Nombre: Santiago Ernesto Pacheco Holguin
# Especialidad: Inteligencia Artificial
# CBTIS 128
# ============================================

print("===================================")
print("   SISTEMA INTELIGENTE DE IBAN")
print("===================================")

opcion = "si"

while opcion == "si":

    iban = input("Ingresa un IBAN: ")

    longitud = len(iban)

    print("\nAnalizando IBAN...\n")

    # Detectar país
    if iban[0:2] == "ES":
        print("País detectado: España")

    elif iban[0:2] == "MX":
        print("País detectado: México")

    elif iban[0:2] == "DE":
        print("País detectado: Alemania")

    elif iban[0:2] == "FR":
        print("País detectado: Francia")

    else:
        print("País no reconocido")

    # Validar longitud
    if longitud < 15:
        print("El IBAN es demasiado corto")

    elif longitud >= 15 and longitud <= 34:
        print("La longitud del IBAN parece válida")

    else:
        print("El IBAN es demasiado largo")

    # Validación básica
    if iban[0:2] == "MX" and longitud == 18:
        print("IBAN mexicano válido")

    elif iban[0:2] == "ES" and longitud == 24:
        print("IBAN español válido")

    else:
        print("No coincide con reglas básicas")

    print("\nConsulta terminada")

    opcion = input("\n¿Deseas analizar otro IBAN? (si/no): ")

print("\nPrograma finalizado")
print("Programa realizado por Santiago Pacheco")
```
